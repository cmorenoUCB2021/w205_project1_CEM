# Project 1: Lyft Bay Wheels - Bike Ridership Strategies

## Executive Summary
The purpose of this report is to provide recommendations to **Lift Bay Wheels** for **"Increasing Bike Ridership in San Francisco"**. The recommendations are supported by an analysis of a dataset (bigquery-public-data.san_francisco.bikeshare) including 983,648 "Bicycle Trips" from August 2013 to August 2016, among 74 bike stations. The bike stations are located in San Francisco (37 Stations), San Jose (18 Stations), Mount View (7 Stations), Redwood City (7 Stations), and Palo Alto (5 Stations).  With 50% of the bike stations, San Francisco is the most important market, as it represents 90.6% of all trips.  There are two types of subscribers reflected in this data set: (1) **Subscribers** who are customers with annual or 30-day membership which represent 86% (846,839) of total trips, and (2) **Customers** who are riders with a 24-hour or 3-day member pass, which represent 14% (136,809) of total trips.  

For the purpose of this report, **Commuter** riders are defined as: "**Subscribers who use a bicycle to go to work (typically in the morning), and to come back home (typically in the afternoon/evening) during weekdays, with a ride duration of 30 minutes or less**". The predominant behavior patter identified for **Subscribers** is aligned with this definition. 93.3% of Subscribers' trips (790,337) take place during weekdays, and 99.4% of weekday rides (785,771) are trips of 30 minutes or less. These 785,771 trips are evenly split between morning and afternoon trips, as 45% of these trips occur before noon, and 55% in the afternoon. This suggests that **Subscribers** are mainly using the bicycle to commute to work. Our analysis indicates that most morning trips occur between 6 and 10 am, and the afternoon/evening trips occurs between 4 and 6 pm.  It is important to note that the average length of commuter's trips is only around 9 minutes. The commuter rider segment is then estimated at 785,711 trips (Subscribers trips that take place on weekday (to commute to work) that last 30 minutes or less).

There are 1,756 routes (**route** is a collection of one or more trips between two stations) in this dataset. When exploring the top seven commuter routes (pair of stations) in the morning and top seven commuter routes in the afternoon, these 14 routes represent 29,122 trips (8% of total commuter trips).  These trips align to the "Commuter" patter described above, as five of the top routes in the morning (50-61, 69-65, 70-55, 70-51, 77-64), have a reciprocal route in the top seven in the afternoon (61-50, 65-69, 55-70, 51-70 and 64-77).  These routes include eight bike stations (50, 51, 55, 60, 61, 65, 69, 70, 74) which (except for station 51) are all in the top 10 stations (identified by volume of total in and out trips).  These stations represent 324,409 commuter trips, or 41.29 % of total commuter trips (when including all trips in and out of the stations).  Thus, by enhancing services in these eight stations, there is the potential to positively impact the customer experience of a large and important segment of the commuters (41.29% of the commuter business).    

**Recommendations:**   
Based on the above analysis, it is recommended to focus on strengthening the **commuter** segment.  The strategy for growing ridership should focus in the following two pillars:  
(1) **Strengthen user experience of current commuters** - While the Subscriber size as percentage of total trips has increased to represent 89% of all trips, its growth has not fully compensated the decline in "Customer’s trips. Some recommendations for strengthening commuter's loyalty are:  
- **Modify or Stop:** While the most frequent length of Commuter trip is 30 minutes or less, the average length of the commuter's rides is 9 minutes. Thus the 45 minutes offer for Subscribers is rarely used, and it may not be valued by them. It is recommended to test alternative offers to the 45 minutes rides offered to Subscribers.  
- **Potential New Alternative:** As the 45 minutes rides for Subscribers may not be of value to them, new services for their commute between 6 to 9 am and 4 to 6pm may be appreciated, as these are the peak time for their commute.  For example, based on reward points, commuters may be offered a free coffee of bottle of water available in key bike stations where the main commute routes take place. The main commute morning routes are 50-61, 69-65, 70-55, 70-51, 77-64, which have reciprocal route in the afternoon. Our team is ready to suppor the testing and evaluation of this and other alternatives.
- **Continue to strengthen and evaluate** - it is not clear if current long term loyalty programs such as partnerships with reward programs with hotels (Hilton), airlines (Delta), payment rewards (Chase & Mastercard) are working. As a follow up to this project, we suggest evaluating how these programs are being used by Subscribers, and how they can be strengthen, as loyalty/reward programs may be important for retaining commuters.

(2) **Expand the commuter customer base** Leverage positive experience of current Subscribers to bring new subscribers. Our analysis indicates that there seems to be a decline of overall trips in 2016 versus 2015, which suggest the need to expand the commuter customer base more aggressively.  Some recommendations for your consideration are:
- **Potential Opportunity:** As top five trips in the morning (50-61, 69-65, 70-55, 70-51, 77-64) have a reciprocal top trip in the afternoon (61-50, 65-69, 55-70, 51-70 and 64-77), it is clear that commuters are using the bicycle to commute to work.  Thus, it is recommended to strengthen corporate programs with companies near the destination stations for the top morning trips.  These bike stations are: 61, 65, 66, 51 and 64. When adding up all the trips (in and out) to these stations, they represent 14% of all commuter trips.
- **Leverage Summer Time:** As there is some seasonality to the business with the summer months (April to August) having the most rides, it is recommended to target these months for special offers to bring new commuters. Leveraging positive commuter experience, we could test programs motivating/compensating commuters for attracting new members (commuters) during summer. In addition, there is opportunity to implement special signing offers to new commuters during summer. Our team is ready to support the definition and analysis of these alternatives.
- **Leverage New Year Resolution wave:** Due to the seasonality of the business, the number of trips start to decline in September to reach the lowest point in December. However, ridership starts to increase again in January. Thus, there may be an opportunity to leverage the current "Gift Cards" program for Christmas to motivate current commuters to bring new commuters for the beginning of the year ("New Year Resolutions").

#### Load Data:
The following section of the report runs all SQL queries, and load results in dataframes which would be used to generate tables and graphs in support of this report.

**(1) SQL QUERY - TO EXTRACT ALL RYDES BETWEEN PAIR OF STATIONS (ROUTES) CLASSIFIED BY:**  
- DAY OF THE WEEK:  Weekday vs. Weekend
- Extract Year, Month, and Hour to analyze seasonality
- TIME OF THE DAY: Afternoon vs. Before_Noon
- Duration of trip: <=30, 30-60, 60-90, 90-120, >=120 minutes duration  

**Note:** JOIN bikeshare_trips and bikeshare_stations to add Landmark to the generated table

````python
%%bigquery df_rides
SELECT 
    EXTRACT(YEAR FROM start_date) as year, EXTRACT(MONTH FROM start_date) as month,
    subscriber_type, start_station_id, end_station_id, 
    (case when (EXTRACT(DAYOFWEEK FROM start_date) > 1) and (EXTRACT(DAYOFWEEK FROM start_date) < 7) then "Weekday"
          else 'Weekend'
     end) as Week_Split,
    (case when EXTRACT(HOUR FROM start_date) >= 12 then "Afternoon"
          else 'Before_Noon'
     end) as Time_Day,
    (case when duration_sec/60 <= 30 then 'a <=30'
          when duration_sec/60 >30  and duration_sec/60 <= 60 then 'b (30-60]'
          when duration_sec/60 >60  and duration_sec/60 <= 90 then 'c (60-90]'
          when duration_sec/60 > 90  and duration_sec/60 <= 120 then 'd (90-120]'
          else 'e >120'
    end) as trip_duration,
    CONCAT('Pair:',CAST(start_station_id AS STRING), '-', CAST(end_station_id AS STRING)) as pair_station_join,
    count(distinct trip_id) as num_trips,
    ROUND(AVG(duration_sec)/60,1) as average_duration,
    ROUND(SUM(duration_sec)/60,1) as total_duration,
FROM 
    (SELECT trip_id, duration_sec, start_date, start_station_id, end_date, end_station_name, end_station_id, bike_number, subscriber_type, landmark
    FROM `bigquery-public-data.san_francisco.bikeshare_trips` JOIN `bigquery-public-data.san_francisco.bikeshare_stations` ON start_station_id = station_id)
GROUP BY 
    year, month, subscriber_type, pair_station_join, Week_Split, trip_duration, start_station_id,end_station_id, Time_Day
ORDER BY 
    year, month, Week_Split, Time_Day, trip_duration, subscriber_type;
````

**(2) SQL query: extracting information by hour of the day**.  Used to explore daily patterns for customers.
- Hour of the day:  1 to 24 hour  
- Including year, month, subscriber_type and Week_split  

````python
%%bigquery dft_nc
SELECT 
    EXTRACT(YEAR FROM start_date) as year,
    EXTRACT(MONTH FROM start_date) as month,
    EXTRACT(HOUR FROM start_date) as hour,
    subscriber_type, start_station_id, end_station_id, 
    (case when (EXTRACT(DAYOFWEEK FROM start_date ) > 1) and (EXTRACT(DAYOFWEEK FROM start_date) < 7) then "Weekday"
          else 'Weekend'
     end) as Week_Split,
    CONCAT('Pair:',CAST(start_station_id AS STRING), '-', CAST(end_station_id AS STRING)) as pair_station_join,
    count(distinct trip_id) as num_trips,
    ROUND(AVG(duration_sec)/60,1) as average_duration,
    ROUND(SUM(duration_sec)/60,1) as total_duration,
FROM 
    `bigquery-public-data.san_francisco.bikeshare_trips`
GROUP BY 
    year, month, hour, subscriber_type, Week_Split,start_station_id, end_station_id
ORDER BY 
    year, month, hour, Week_Split;
````

**(3) SQL QUERY:** to extract the name and region (landmark) of the bike station.  
This query is used to identify name and landmark associated to key bike stations.

````python
%%bigquery df_stations
SELECT *
FROM `bigquery-public-data.san_francisco.bikeshare_stations`
````

**(4) SQL QUERY: To gather top 20 pair of stations (trips) for morning and afternoon commute for trips of 30 minutes or less.**  
The query also split information by: 
- DAY OF THE WEEK:  Weekday vs. Weekend
- Extract Year, Month, and Hour to analyze seasonality
- TIME OF THE DAY: Afternoon vs. Before_Noon
- Duration of trip: <=30, 30-60, 60-90, 90-120, >=120 minutes duration

````python
%%bigquery df_pairs
SELECT *
FROM
    (SELECT 
        landmark,Time_Day,trip_duration, subscriber_type,
        CONCAT('Pair:',CAST(start_station_id AS STRING), '-', CAST(end_station_id AS STRING)) as pair_station_join,
        count(distinct trip_id) as num_trips,
        ROUND(AVG(duration_sec)/60,1) as average_duration,
        ROW_NUMBER() OVER (PARTITION BY Time_Day ORDER BY count(distinct trip_id) DESC) AS num_trip_rank
    FROM 
        (SELECT trip_id, duration_sec, start_date, start_station_id, end_date, end_station_name, end_station_id, bike_number, subscriber_type, landmark,
                (case when duration_sec/60 <= 30 then 'a <=30'
                      when duration_sec/60 >30  and duration_sec/60 <= 60 then 'b (30-60]'
                      when duration_sec/60 >60  and duration_sec/60 <= 90 then 'c (60-90]'
                      when duration_sec/60 > 90  and duration_sec/60 <= 120 then 'd (90-120]'
                      else 'e >120'
                end) as trip_duration,
                (case when EXTRACT(HOUR FROM start_date) >= 12 then "Afternoon"
                      else 'Before_Noon'
                end) as Time_Day,
         FROM `bigquery-public-data.san_francisco.bikeshare_trips` JOIN `bigquery-public-data.san_francisco.bikeshare_stations` ON start_station_id = station_id)
         WHERE trip_duration = 'a <=30' and subscriber_type = "Subscriber"
         GROUP BY landmark, subscriber_type, pair_station_join, Time_Day, trip_duration
         ORDER BY Time_Day, num_trips, num_trip_rank DESC)
WHERE
      num_trip_rank <= 20
ORDER BY Time_Day, num_trip_rank, num_trips DESC, landmark;
````

**(5) SQL QUERY** - Information by Region (landmark) - including number of trips, number of bike stations,
average trips per bike station, and calculate percentage of totals for number of trips and number of stations.

````python
%%bigquery df_regions
SELECT landmark, count(trip_id) as num_trips, count(distinct start_station_id) as num_stations, 
       ROUND(IEEE_DIVIDE(count(trip_id),count(distinct start_station_id)),0) as avg_trip_by_station,
       ROUND(count(trip_id)/SUM(count(trip_id)) OVER () *100, 1) as percent_tot_trips,
       ROUND(count(distinct start_station_id)/SUM(count(distinct start_station_id)) OVER () *100, 1) as percent_tot_stations
FROM
    (SELECT trip_id, duration_sec, start_date, start_station_id, end_date, end_station_name, end_station_id, bike_number, subscriber_type, landmark
    FROM `bigquery-public-data.san_francisco.bikeshare_trips` JOIN `bigquery-public-data.san_francisco.bikeshare_stations` ON start_station_id = station_id)
GROUP BY 
    landmark
ORDER BY 
    num_trips DESC;
````

**(6) SQL QUERY:** to get top routes (pair of stations) based on total trips in the morning and afternoon.

````python
%%bigquery df_pairtotal
#Getting top pairs (morning and afternoon)
    SELECT 
        landmark,trip_duration, subscriber_type,
        CONCAT('Pair:',CAST(start_station_id AS STRING), '-', CAST(end_station_id AS STRING)) as pair_station_join,
        count(distinct trip_id) as num_trips,
        ROUND(AVG(duration_sec)/60,1) as average_duration,
    FROM 
        (SELECT trip_id, duration_sec, start_date, start_station_id, end_date, end_station_name, end_station_id, bike_number, subscriber_type, landmark,
                (case when duration_sec/60 <= 30 then 'a <=30'
                      when duration_sec/60 >30  and duration_sec/60 <= 60 then 'b (30-60]'
                      when duration_sec/60 >60  and duration_sec/60 <= 90 then 'c (60-90]'
                      when duration_sec/60 > 90  and duration_sec/60 <= 120 then 'd (90-120]'
                      else 'e >120'
                end) as trip_duration,
           FROM 
               `bigquery-public-data.san_francisco.bikeshare_trips` JOIN `bigquery-public-data.san_francisco.bikeshare_stations` ON start_station_id = station_id)
     WHERE 
        trip_duration = 'a <=30' and subscriber_type = "Subscriber"
     GROUP 
        BY pair_station_join, subscriber_type, trip_duration, landmark
     ORDER 
        BY num_trips DESC;
````

#### (9) Opportunity to focus on the top 10 Bike Stations representing 50% of commuter segment
To better optimize the use of the marketing budget, there is an opportunity to focus on improving services for the top bike stations serving the commuter segment.  The table below presents the top 10 bike stations based on the number of trips (30 minutes or less) from and to the stations (trips leaving and arriving to the station).  These bike stations are all located in San Francisco, and they represent 50 percent of all commuter trips.  There is an opportunity to optimize investment in these stations, in order to strengthen the commuter's customer experience.

````python
top_num = 10
stations_array = np.zeros((df_rides['start_station_id'].max()+1,2)).astype(int)
df_rides_subs_30min = df_rides[(df_rides['subscriber_type'] == 'Subscriber') & (df_rides['trip_duration'] == 'a\xa0<=30')]
for index, row in df_rides_subs_30min.iterrows():
    stations_array[row["start_station_id"]][0] += row["num_trips"]
    stations_array[row["end_station_id"]][1] += row["num_trips"]

df_stations_load = pd.DataFrame({"station_id": range(df_rides['start_station_id'].max()+1),
                                 "Trips Leaving Station":stations_array[:,0],
                                 "Trips to Station":stations_array[:,1],
                                })

df_stations_load["Total Trips"] = df_stations_load["Trips Leaving Station"] + df_stations_load["Trips to Station"] 
df_stations_load.sort_values(by=["Total Trips"], ascending=False)

df_stations_join = pd.merge(df_stations_load, df_stations, on='station_id', how='left')
df_stations_join = df_stations_join.dropna()

dfs_sorted = df_stations_join.sort_values(by=["Total Trips"], ascending=False)
dfs_sorted_top = dfs_sorted.iloc[0:top_num,][["station_id","Trips Leaving Station",
                                            "Trips to Station", "Total Trips", "name", "landmark"]]

tt_leaving_station = dfs_sorted_top["Trips Leaving Station"].sum()

tt_to_station = dfs_sorted_top["Trips to Station"].sum()

dfs_sorted_top.index = range(1,top_num+1)
dfs_sorted_top
````

````python
print("Total Trip Leaving top ",top_num," stations: ",tt_leaving_station, ", which represents: ", 
      round((tt_leaving_station/total_cummuter_trips)*100,1),"% of all commuter trips")
print("Total Trip to top ",top_num, " stations: ",tt_to_station, ", which represents: ", 
      round((tt_to_station/total_cummuter_trips)*100,1),"% of all commuter trips")
````

````python
def get_stations (data):
    x = []
    for pair in data:
        if int(pair.split("-")[1]) not in x:
            x.append(int(pair.split("-")[1]))
        if int(pair.split("-")[0].split(":")[1]) not in x:
            x.append(int(pair.split("-")[0].split(":")[1]))
    return(sorted(x))
````

## Summary: Key Findings
The dataset includes 983,648 bicycle trips among 74 bike stations which took place between August 2013 and August 2016. Some of the key findigs are:
* There are five cities (landmarks) included in the dataset (San Francisco (37 Stations), San Jose (18 Stations), Mount View (7 Stations), Redwood City (7 Stations), and Palo Alto (5 Stations).  With 50% of the bike stations, San Francisco represent 90.6% of all trips, with the highest average number of trips per station at 24,087 trips/station.

* Subscribers are customers with annual or 30-day membership and represent 86% (846,839) of total trips, and Customers are riders with a 24-hour or 3-day member pass, which represent 14% (136,809) of total trips.  In the last two years (2015 and 2016), "Subscribers" have increased as percentage of total riders to represent 89% of total riders.  However, their increase has not compensated the lost of trips related to "Costumers".  This is reflected in a sligth decrease of ridership in 2016 versus 2015. 
* There is some seasonality to the business with a peak in the summer time (April to August), and a decline by the end of the year, starting in September.  Thus, activities to attract new customers in the summer time are likely to have the greatest impact.  
* For the purpose of this report, Commuter Riders are defined as: "Subscribers who use a bicycle to go to work (typically in the morning), and to come back home (typically in the afternoon/evening) during weekdays, with a trip duration of 30 minutes or less". The predominant behavioral patter identified for Subscribers is aligned with this definition. 93.3% of Subscriber's trips (790,337) take place during weekdays, and 99.4% of weekday rides (785,771) are trips of 30 minutes or less. Thus, the commuter segment is estimated at 785,771 trips (80% of all trips (983,648).

### Questions of your own
**Question 1:**   
What is the largest landmarks (markets) as measured by number of trips? How many bike stations the largest market has? what is the average number of trips for the largest market?  
**Answer:**   
The largest market (landmark) is San Francisco with 891,223 trips which represents 90.6% of all trips.  San Francisco has 37 (out of 74) bike stations (50% of all bike stations).  Being the largest market, the average number of trips per station in San Franciso is 24,087 which is 7 times larger than the next landmark (Mountain View).  The table below presents a summary of trips and stations by landmark (market).

````python
%%bigquery landmarks_table
SELECT   landmark,
         Count(trip_id)                                                                                AS num_trips,
         Count(DISTINCT start_station_id)                                                              AS num_stations,
         Round(Ieee_divide(Count(trip_id),Count(DISTINCT start_station_id)),0)                         AS avg_trip_by_station,
         Round(Count(trip_id)                  /Sum(Count(trip_id)) OVER () *100, 1)                   AS percent_tot_trips,
         Round(Count(DISTINCT start_station_id)/Sum(Count(DISTINCT start_station_id)) OVER () *100, 1) AS percent_tot_stations
FROM     (
                SELECT trip_id,
                       duration_sec,
                       start_date,
                       start_station_id,
                       end_date,
                       end_station_name,
                       end_station_id,
                       bike_number,
                       subscriber_type,
                       landmark
                FROM   `bigquery-public-data.san_francisco.bikeshare_trips`
                JOIN   `bigquery-public-data.san_francisco.bikeshare_stations`
                ON     start_station_id = station_id)
GROUP BY landmark
ORDER BY num_trips DESC;
````

**Question 2: What is the most common trip duration during weekdays and weekends for Subscribers? Is this different for Customers?**    
**Answer:**  
The most frequent trip for subscriber during weekdays and weekends is trips of 30 minutes or less.
These trips represent 99.4% of all Subscribers trips (841,515/846,839).
While these trips are classified as being of 30 minutes or less, their average duration is only 8 to 9 minutes.
It is likely that Subscribers are communiting from home to work and back home using these short trips.
The table below presents details of split for trips (<=30 minutes, 30 to 60 minutes, 60 to 90 minutes,
90 to 120 minutes and more than 120 minutes).
For customers, trips of 30 minutes or less are the most common at 94,796, which represent 69% of al "Customer Trips".
However, Subscribers is much larger segment of riders, with 841,515 trips of 30 minutes or less, versus
only 94,796 trips for "Customers".

````python
%%bigquery duration_subscriber
SELECT 
      subscriber_type,
      (case when duration_sec/60 <= 30 then 'a <=30'
            when duration_sec/60 >30  and duration_sec/60 <= 60 then 'b (30-60]'
            when duration_sec/60 >60  and duration_sec/60 <= 90 then 'c (60-90]'
            when duration_sec/60 > 90  and duration_sec/60 <= 120 then 'd (90-120]'
            else 'e >120'
      end) as trip_duration,
      (case when (EXTRACT(DAYOFWEEK FROM start_date) > 1) and (EXTRACT(DAYOFWEEK FROM start_date) < 7) then "Weekday"
            else 'Weekend'
      end) as Week_Split,
      count(distinct trip_id) as num_trips,
      ROUND(AVG(duration_sec)/60,1) as average_duration
FROM 
    `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE 
    subscriber_type = "Subscriber"
GROUP BY 
    subscriber_type, trip_duration, Week_Split
ORDER BY 
    Week_Split,subscriber_type, trip_duration;
````

````python
duration_subscriber
````

````python
%%bigquery duration_customer
SELECT 
      subscriber_type,
      (case when duration_sec/60 <= 30 then 'a <=30'
            when duration_sec/60 >30  and duration_sec/60 <= 60 then 'b (30-60]'
            when duration_sec/60 >60  and duration_sec/60 <= 90 then 'c (60-90]'
            when duration_sec/60 > 90  and duration_sec/60 <= 120 then 'd (90-120]'
            else 'e >120'
      end) as trip_duration,
      (case when (EXTRACT(DAYOFWEEK FROM start_date) > 1) and (EXTRACT(DAYOFWEEK FROM start_date) < 7) then "Weekday"
            else 'Weekend'
      end) as Week_Split,
      count(distinct trip_id) as num_trips,
      ROUND(AVG(duration_sec)/60,1) as average_duration
FROM 
    `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE 
    subscriber_type = "Customer"
GROUP BY 
    subscriber_type, trip_duration, Week_Split
ORDER BY 
    Week_Split,subscriber_type, trip_duration;
````

### Bonus activity queries (optional - not graded - just this section is optional, all other sections are required)

The bike share dynamic dataset offers multiple tables that can be joined to learn more interesting facts about the bike share business across all regions. These advanced queries are designed to challenge you to explore the other tables, using only the available metadata to create views that give you a broader understanding of the overall volumes across the regions(each region has multiple stations)

- **Top 5 popular station pairs in each region**

````python
%%bigquery popular_station_pairs_by_region
SELECT *
FROM
    (SELECT 
    landmark,
    CONCAT('Pair:',CAST(start_station_id AS STRING), '-', CAST(end_station_id AS STRING)) as pair_station_join,
    count(distinct trip_id) as num_trips,
    ROUND(AVG(duration_sec)/60,1) as average_duration,
    ROW_NUMBER() OVER (PARTITION BY landmark ORDER BY count(distinct trip_id) DESC) AS num_trip_rank
    FROM 
        (SELECT trip_id, duration_sec, start_date, start_station_id, end_date, end_station_name, end_station_id, bike_number, subscriber_type, landmark
        FROM `bigquery-public-data.san_francisco.bikeshare_trips` JOIN `bigquery-public-data.san_francisco.bikeshare_stations` ON start_station_id = station_id)
    GROUP BY landmark, pair_station_join
    ORDER BY num_trips, num_trip_rank DESC)
WHERE
    num_trip_rank <= 5
ORDER BY 
    landmark, num_trip_rank, num_trips DESC;
````
